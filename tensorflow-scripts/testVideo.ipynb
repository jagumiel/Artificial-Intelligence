```python
import cv2
import numpy as np
import re

from PIL import Image
from PIL import ImageDraw
from PIL import ImageColor
from PIL import ImageFont

vidcap = cv2.VideoCapture('video_nurbu.mp4')
success,imagenBruta = vidcap.read()

while success:
    #resize image
    img = cv2.resize(imagenBruta,(320, 320))
    input_data = np.expand_dims(img, axis=0)
    image = Image.fromarray(imagenBruta)
    
    draw = ImageDraw.Draw(image)
    im_w, im_h = image.size
    print(image.size)
    
    cv2.imshow('Video Test', np.array(image, dtype = np.uint8))
    #Wait. Also exits when q key is pressed. Needed to show the video, DO NOT DELETE.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    success,imagenBruta = vidcap.read()
    
    
vidcap.release()
cv2.destroyAllWindows()
```
